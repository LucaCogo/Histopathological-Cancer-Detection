## Import packages

```python
# per caricamento file csv
import pandas as pd
# per resampling 
from sklearn.utils import resample
# per spostamento file
import shutil
```

## Resizing training data
Carico il file .csv contenente le informazioni su tutte le immagini, estraggo quindi un sottoinsieme delle immagini in modo randomico per ogni classe. Copio le immagini estratte nella nuova cartella.

```python
# carico il file contenente il riferimento alla label per ogni immagine
df_label = pd.read_csv('./train_labels.csv')

# visualizzo il bilanciamento delle classi
print('<---------- Bilanciamento delle classi: ---------->')
print(df_label['label'].value_counts())
```

Siccome il dataset è fortemente sbilanciato estraggo due sotto dataset, uno per ogni classe. E a partire da questi in modo casuale estraggo lo stesso numero di classi, così facendo mi assicuro che vi sia bilanciamento nel dataset finale.

```python
# dato che le classi sono sbilanciate divido il dataset per avere le due classi divise
df_negative_class = df_label[df_label['label'] == 0]
df_positive_class = df_label[df_label['label'] == 1]

# estraggo 12544 esempi da ogni classe (in questo modo ho 25088 esempi totali)
df_pos = resample(df_positive_class, replace=False, n_samples=12544, random_state=10)
df_neg = resample(df_negative_class, replace=False, n_samples=12544, random_state=10)

# unisco i dati sulle label delle immagini estratte e scrivo tutto su un csv
pd.concat([df_pos, df_neg]).to_csv("train_small_label.csv", index=False)
```

A questo punto ho a disposizione il filename di 25088 immagini contenenti esempi per entrambe le classi. Copio quindi queste immagini estratte randomicamente dalla cartella contenente tutte le $220$k immagini alla nuova cartella ridotta.

```python
# path in cui sono state salvate tutte le 220k immagini
global_dir_path = 'C://Users//Riccardo//Downloads//train/'

# salvo la classe positiva (1), contenente i tumori nella cartella train_small/1/
resized_dir_path = 'C://Users//Riccardo//Downloads//train_small/1/'

# itero su tutte le immagini con tumore (class = 1)
for _, row in df_pos.iterrows():
    # estraggo il nome del file
    filename = row['id'] + '.tif'

    # copio il file nella nuova directory
    shutil.copy2(global_dir_path + filename, resized_dir_path + filename)
```

```python
# salvo la classe negativa (0), nella cartella train_small/0/
resized_dir_path = 'C://Users//Riccardo//Downloads//train_small/0/'
# itero su tutte le immagini senza tumore (class = 0)
for _, row in df_neg.iterrows():
    # estraggo il nome del file
    filename = row['id'] + '.tif'

    # copio il file nella nuova directory
    shutil.copy2(global_dir_path + filename, resized_dir_path + filename)
```
